### <span style="color:#F05650">학교나 어린이 보육시설에 인접한 지역에 대피소가 지정될 것이다.</span>

설명<br>
어린이와 학생들은 재난 발생 시, 보호자와 함께 대피할 필요가 있으며, 해당 가설은 이러한 필요를 충족시킬 수 있을 것으로 예상된다.  이와 반대로 고령자의 경우도 포함될 수 있음.

계획<br>
각 학교와 대피소 사이의 거리 계산<br>
거리가 가까운 대피소 일수록, 더 높은 가중치 부여

### 라이브러리

```python
import pandas as pd
```

```python
school_data = pd.read_csv('../Data/School/전국초중등학교_위치데이터.csv', encoding='cp949')
```

```python
school_data.head()
```

```python
print('-'*10,'info','-'*10)
print(school_data.info())
print('-'*10,'describe','-'*10)
print(school_data.describe())
print('-'*10,'null','-'*10)
print(school_data.isnull().sum())
```

부산광역시에서 운영중인 상태의 학교만 출력

```python
school_data = school_data[(school_data['시도교육청명'].str.contains('부산광역시')) & (school_data['운영상태'] == '운영')]
```

```python
school_data.head()
```

필요없는 컬럼 삭제

```python
school_data.drop(columns=school_data.columns[3:6], inplace=True)
school_data.drop(columns=['학교ID','시도교육청코드','시도교육청명','교육지원청코드','교육지원청명','생성일자','변경일자','데이터기준일자','운영상태'], inplace=True)
```
